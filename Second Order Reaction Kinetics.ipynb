```python
#Second Order Reaction Kinetics
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def reacnsecond_order(c,t):
    ca=c[0]
    cb=c[1]
    dcadt = -0.3*ca*cb
    dcbdt = -0.3*ca*cb
    return [dcadt,dcbdt]
```

```python
test=reacnsecond_order([0.2,0.3],0)
```

```python
test
```

```python
c0=[0.2,0.3]
```

```python
t=np.linspace(0,5,20)
```

```python
c=odeint(reacnsecond_order,c0,t)
```

```python
plt.xlabel('t min')
plt.ylabel('ca, cb')
plt.plot(t,c)
plt.grid()
plt.legend(['ca', 'cb'])
```
